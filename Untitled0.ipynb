Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use,and performance.

Answer:- NLTK (Natural Language Toolkit) and spaCy are two popular Python libraries used for Natural Language Processing (NLP). Both provide tools for working with text data, but they differ in design, speed, and typical use cases.

1. Features

* NLTK:- One of the oldest NLP libraries.

=> Provides many tools for tokenization, stemming, tagging, parsing, and classification.
=> Includes many datasets and educational resources.
=> Often used for research and teaching NLP concepts.

* spaCy:-

Modern industrial-grade NLP library.

Provides tokenization, part-of-speech tagging, named entity recognition (NER), dependency parsing, and word vectors.

Designed for real-world production applications.

2. Ease of Use

* NLTK:- Easy for beginners who want to learn NLP step-by-step.
=> However, many tasks require multiple lines of code and manual steps.

* spaCy:- Easier for practical applications because it provides ready-to-use pipelines.
+> Requires less code for complex tasks like entity recognition.

3. Performance

* NLTK:- Slower because it is written mainly in pure Python.
=> Better suited for learning and experimentation.

* spaCy:- Much faster and more efficient.

=> Implemented using Cython and optimized algorithms, making it suitable for large-scale NLP systems.

Question 2:-What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?

Answer:

TextBlob is a Python library used for processing textual data and performing common Natural Language Processing (NLP) tasks. It provides a simple API that allows developers to perform complex NLP operations easily. TextBlob is built on top of NLTK and Pattern libraries.

TextBlob can perform sentiment analysis by determining whether a text expresses positive, negative, or neutral sentiment. It provides two values: polarity and subjectivity. Polarity ranges from -1 to +1 and indicates whether the sentiment is negative or positive.

It also supports translation between languages using simple commands. In addition, TextBlob can perform part-of-speech tagging, which identifies the grammatical role of each word such as nouns, verbs, and adjectives.

Another important feature of TextBlob is tokenization, which allows the text to be broken into sentences or words so that it can be analyzed more effectively.

Question 4:- Describe the architecture and functioning of a Recurrent Neural Network (RNN).

Answer:

A Recurrent Neural Network (RNN) is a type of neural network designed for sequential data such as text, speech, and time-series data. Unlike traditional neural networks, RNNs have the ability to remember previous inputs while processing new data.

The architecture of an RNN consists of three main layers: input layer, hidden layer, and output layer. The input layer receives the sequence of data such as words in a sentence. The hidden layer processes the input and also stores information from previous time steps. This stored information is called the hidden state and acts as memory for the network. The output layer generates the final prediction.

The functioning of an RNN involves processing one element of the sequence at a time. At each step, the network takes the current input and combines it with the previous hidden state to produce a new hidden state. This allows the network to capture contextual information and relationships between elements in the sequence.

However, RNNs can face problems such as the vanishing gradient problem, which makes it difficult for the network to learn long-term dependencies.

Question 5:- What is the key difference between LSTM and GRU networks in NLP applications?

Answer:

LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are advanced versions of Recurrent Neural Networks designed to overcome the limitations of traditional RNNs, particularly the vanishing gradient problem.

LSTM networks use three gates: input gate, forget gate, and output gate. These gates control how information flows through the network and help it remember or forget information over long sequences. Because of this structure, LSTM models can capture long-term dependencies in sequential data.

GRU networks are a simplified version of LSTM. They combine the input and forget gates into a single update gate and use a reset gate to control memory. Because GRUs have fewer gates and parameters, they are simpler and faster to train compared to LSTM models.

Both models are widely used in NLP tasks such as text classification, language translation, and speech recognition.

Question 6:-Write a Python program using TextBlob to perform sentiment analysis on the given paragraph.

```python
from textblob import TextBlob

text = """
i had a great experience using the new mobile banking app.
the interface is intuitive and customer support was quick to resolve my issue.
however the app did crash once during a transaction which was frustrating
"""

blob = TextBlob(text)

print("polarity:", blob.sentiment.polarity)
print("subjectivity:", blob.sentiment.subjectivity)
```

Question 7:- Perform string tokenization and frequency distribution using Python and NLTK.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """natural language processing is a fascinating field that combines linguistics
computer science and artificial intelligence"""

tokens = word_tokenize(text)

fdist = FreqDist(tokens)

print(tokens)
print(fdist)
```

Question 8:-Implement a basic LSTM model in Keras for text classification?

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
    "i love this app",
    "this app is amazing",
    "i hate this app",
    "this app is bad"
]

labels = np.array([1,1,0,0])

tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

padded = pad_sequences(sequences)

model = Sequential()
model.add(Embedding(100,16,input_length=padded.shape[1]))
model.add(LSTM(32))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(padded,labels,epochs=10)
```

Question 9:-Using spaCy, build a simple NLP pipeline that includes tokenization, lemmatization, and entity recognition.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
homi jehangir bhaba was an indian nuclear physicist who played a key role
in the development of india's atomic energy program
"""

doc = nlp(text)

print("tokens and lemmas")
for token in doc:
    print(token.text, token.lemma_)

print("\nnamed entities")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
